```python
import numpy as np
import math
import matplotlib.pyplot as plt

m=1
def monteCarlo(d):  
    n=1000000
    rnd = (np.random.random(size=[n,math.ceil(m*d)]))/(m**0.5)
    distances = np.sum(rnd**2, axis=1)**0.5
    
    return (2**d)*np.mean(distances < 1)

# Create x values from 0 to 50
x = np.arange(0, 20, 1/m) 
y=[monteCarlo(d) for d in  x/m**0.5]
z=[np.pi**(d/2)/math.gamma(d/2+1) for d in x]
    


# Calculate corresponding y values using the volume function

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x, z, 'go-', label='Gamma')
plt.plot(x, y, 'ro', label='Monte-Carlo')
plt.xlabel('Dimensions')
plt.ylabel('Volume')
plt.title('Volume Function')
plt.grid(True, which='major', linestyle='-', alpha=0.8)
plt.grid(True, which='minor', linestyle='-', alpha=0.8)
plt.minorticks_on()
plt.legend()

# Add some padding to the y-axis to better show the curve
plt.margins(y=0.1)

plt.show()
```
